## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](https://github.com/shinonome-inc/PythonTutorialForDSCourse/blob/main/exercise/images/plot_1.png?raw=1)

```python
data = pd.read_csv('/content/hist_data.csv')
x1 = data['aデータ']
x2 = data['bデータ']
plt.hist(x1, bins=30, label='aデータ', color='gold', alpha=0.5)
plt.hist(x2, bins=30, label='bデータ', color='aqua', alpha=0.5)
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](https://github.com/shinonome-inc/PythonTutorialForDSCourse/blob/main/exercise/images/plot_2.png?raw=1)

```python
data = pd.read_csv('/content/scatter_data.csv')
data_length=data["length (cm)"]
data_width=data["width (cm)"]
data_class=data["class"]
plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.xlabel("length(cm)")
plt.ylabel("width(cm)")
plt.scatter(data_length[data_class==0], data_width[data_class==0], color='red', label='class0')
plt.scatter(data_length[data_class==1], data_width[data_class==1], color='blue', label='class1')
plt.scatter(data_length[data_class==2], data_width[data_class==2], color='green', label='class2')
plt.legend()
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](https://github.com/shinonome-inc/PythonTutorialForDSCourse/blob/main/exercise/images/plot_3_1.png?raw=1)

![問3_2](https://github.com/shinonome-inc/PythonTutorialForDSCourse/blob/main/exercise/images/plot_3_2.png?raw=1)

```python
data = pd.read_csv('line_data.csv')
x = data['期間']
y = data['費用']
z = data['件数']

plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.ylabel('件数')
plt.plot(x, z, marker='o', linestyle='-', color='blue')
plt.grid(True)
plt.xticks(rotation=45, fontsize=8)
plt.show()
plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.ylabel('費用（円)')
plt.bar(x, y)
plt.grid(True)
plt.xticks(rotation=45, fontsize=8)
plt.show()
```
